```python
#Muse
category = "Engineer"
location = "sf"
page = "0"
descending = "True"
 
# link = "https://jobs.github.com/positions.json?description=" + keyword + "&page=" + page

link = "https://themuse.com/api/public/jobs?category"+category+ "&page=" + page+ "&descending=" + descending

print(link)
 
# https://jobs.github.com/positions.json?description=python&location=sf
```

```python
r = requests.get(link)
```

```python
jobs = json.loads(r.text)
```
